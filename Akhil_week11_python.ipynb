## **Introduction**

In this Housing dataset we have demographic information  (income, population, home occupancy) in neighborhoods, location of neighborhoods (latitude, longitude) and general information homes in neighboring areas (number of rooms, number of bedrooms, age of the house). These statistics are at the regional granularity, they correspond to averages or averages.



## **Variables used in the dataset :**
1. **Longitude:** A measure of how far west a house is; a more negative value is farther west.

2. **Latitude:** 	A measure of how far north a house is; a higher value is farther north.

3. **housing_median_age:** Median age of a house within a block; a lower number is a newer building.

4. **total_rooms:** Total number of rooms within a block.

5. **total_bedrooms:**	Total number of bedrooms within a block.

6. **population:** Total number of people residing within a block

7. **households:** 	Total number of households, a group of people residing within a home unit, for a block.

8. **median_income:** 	Median income for households within a block of houses.

9. **median_house_value:** Median house value for households within a block.

10. **ocean_proximity:** Location of the house w.r.t ocean/sea.



### **Statastical Representation of Data**

Through the satatasticla approch, the columns are devided into Nominal , Discret , Continues , Interval level data.

1. **Nominal level :** ocean_proximity.

2. **Discrete level:** housing_median_age,total_rooms,total_bedrooms,population,households,median_house_value.

3. **Continues level:** median_income.

4. **Interval level:** latitude,longitude.

## **5.Create a data set by filling the missing data with the mean value of the total_bedrooms in the original data set.**

**From the above info we fould the python is reading total_rooms,total_bedrooms, population,households, median_house_value and float but they should be read as Integer.**

## **10. Please create a new column named total_bedroom_size. If size 10 or less. It should be quoted as small. If the size is 11 or more but 1000 less, it should be medium, otherwise it should be large.**

**Not to lose the main dataframe i just copied the whole df to df2 to safety purpose even df2 gose wrong , df will be there.**

```python
#using deep copy functinality to copy whole data into df2
df2 = df.copy()
```

```python
df2.info()
```

```python
df2.head()
```

```python
print(f"max value in total bedrooms is {df2['total_bedrooms'].max()}")
print(f"min value in total bedrooms is {df2['total_bedrooms'].min()}")
```

creating the another colum as total bedrooms size were it is catogorical data and this column is refers from column total_bedrooms  were it value is 10 or less is **small**,11 or more but 1000 less is **medium**, otherwise is **large**

1. the max value we got is 6445.0 because at starting we used mean to fill null value so that the reason got 6445.0 as higher value
2. min value in total bedrooms is 1

```python
cut_bins =[0,10,999,6446] 
cut_total = ['small','medium','large']

df2['total_bedroom_size'] = pd.cut(df2['total_bedrooms'],bins = cut_bins , labels = cut_total)
```

**Above Code Explanation:**
1. I am using pd.cut, Pandas cut() function is used to separate the array elements into different bins.
2. bins are used to divide the value into specific parts.
* if you look in to logic cut_bins have 4 different values in bin they are 0,10,999,6446
it can be writen as (0,10] (10,999] (999,6446]. So , 10 is included in (0,10] but not in (10,999](**It is important to know that ( parenthesis means not included and ] means included**) likewise 0-10 is small , 10-999 is median , greater than 1000  is large.
